```
import numpy as np
import torch
from torch.autograd import Variable
from torch import optim
from torch.utils.data import DataLoader
from torchvision.datasets import MNIST
from torchvision.transforms import transforms

train_dataset = MNIST(root='./data',
                            train=True,
                            transform=transforms.ToTensor(),
                            download=True)

test_dataset = MNIST(root='./data',
                           train=False,
                           transform=transforms.ToTensor())

train_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                                           batch_size=4,
                                           shuffle=True)

test_loader = torch.utils.data.DataLoader(dataset=test_dataset,
                                          batch_size=4,
                                          shuffle=False)

class LogisticRegression(torch.nn.Module):
    def __init__(self):
        super(LogisticRegression,self).__init__()
        self.linear = torch.nn.Linear(28*28,10)
    def forward(self,x):
        out = self.linear(x)
        return out

model = LogisticRegression()
model.cuda()
loss_function = torch.nn.CrossEntropyLoss()
optimizer = optim.Adadelta(model.parameters())
optimizer2 = optim.RMSprop(model.parameters())
optimizer3 = optim.Adam(model.parameters())
optimizer4 = optim.Adagrad(model.parameters())

def train(train_loader,o):
    model.train()
    running_loss = 0
    for bid,(image,label) in enumerate(train_loader):
        image = Variable(image.view(-1,28*28).cuda())
        label = Variable(label.cuda())
        o.zero_grad()
        output = model(image)
        loss = loss_function(output,label)
        running_loss += loss.item()
        loss.backward()
        o.step()
    return running_loss/len(train_loader)

def valid(test_loader):
    model.eval()
    running_loss = 0
    correct = 0
    total = 0
    for bid,(image,label) in enumerate(test_loader):
        image = Variable(image.view(-1,28*28).cuda(),requires_grad=False)
        label = Variable(label.cuda(),requires_grad=False)
        output = model(image)
        loss = loss_function(output,label)
        running_loss += loss
        _,predict = torch.max(output.data,1)
        correct += (predict==label).sum()
        total +=label.size(0)
    val_loss = running_loss/len(test_loader)
    val_acc = float(correct)/total
    return val_loss,val_acc

tll = [[] for i in range(4)]
vll = [[] for i in range(4)]
val = [[] for i in range(4)]

c = 0
for o in [optimizer,optimizer2,optimizer3,optimizer4]:
    for epoch in range(50):
        tl = train(train_loader,o)
        vl,va = valid(test_loader)
        print('epoch %d, loss: %.4f val_loss: %.4f val_acc: %.4f'
              % (epoch, tl, vl, va))
        tll[c].append(tl)
        vll[c].append(vl)
        val[c].append(va)
    c+=1
    print("------------------------")

from matplotlib import pyplot as plt

plt.figure()
plt.plot(range(50), tll[0], 'r-', label='adadelta_tl')
plt.plot(range(50), vll[0], 'g-', label='adadelta_vl')
plt.plot(range(50), tll[1], 'b-', label='rmsprop_tl')
plt.plot(range(50), vll[1], 'c-', label='rmsprop_vl')
plt.plot(range(50), tll[2], 'm-', label='adam_tl')
plt.plot(range(50), vll[2], 'y-', label='adam_vl')
plt.plot(range(50), tll[3], 'k-', label='adagrad_tl')
plt.plot(range(50), vll[3], 'w-', label='adagrad_vl')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.grid()

plt.figure()
plt.plot(range(50), val[0], 'r-', label='adadelta_acc')
plt.plot(range(50), val[1], 'g-', label='rmsprop_acc')
plt.plot(range(50), val[2], 'b-', label='adam_acc')
plt.plot(range(50), val[3], 'c-', label='adagrad_acc')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('acc')
plt.grid()
```
